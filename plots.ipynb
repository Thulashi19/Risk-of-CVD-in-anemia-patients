```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load dataset
df = pd.read_csv('actualdata.csv')

# Filter heart disease patients only
df_hd = df[df['Heartdisease'] == 1].copy()

# Clean Gender column to consistent labels
df_hd['Gender'] = df_hd['Gender'].str.strip().str.capitalize()

# Create color map for genders
color_map = {'M': 'blue', 'F': 'red'}  # blue for Male, orange for Female

plt.figure(figsize=(8, 6))

# Scatter plot: x is Hb level, y is jittered to separate points visually
for gender, color in color_map.items():
    subset = df_hd[df_hd['Gender'] == gender]
    y_jitter = np.random.normal(loc=0, scale=0.1, size=len(subset))  # small vertical jitter
    plt.scatter(subset['Hb'], y_jitter, label=gender, color=color, alpha=0.7, edgecolor='k', s=60)

plt.yticks([])  # Hide y-axis since it's jitter
plt.xlabel('Hemoglobin (Hb) Level (g/dL)', fontsize=14)
plt.title('Hb Distribution of Heart Disease Patients by Gender', fontsize=16, fontweight='bold')
plt.legend(title='Gender')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# RDW-CV plot with jitter
plt.figure(figsize=(8, 6))
color_map = {1: 'red', 0: 'blue'}  # 1 = heart disease, 0 = no heart disease

for hd_status, color in color_map.items():
    subset = df_hd[df_hd['HeartDisease'] == hd_status]
    y_jitter = np.random.normal(loc=0, scale=0.1, size=len(subset))
    plt.scatter(subset['RDW-CV'], subset['HeartDisease'] + y_jitter, 
                alpha=0.6, color=color, label=f'Heart Disease {hd_status}')

plt.xlabel('RDW-CV')
plt.ylabel('Heart Disease Status')
plt.title('RDW-CV vs Heart Disease Status')
plt.yticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('actualdata.csv')

# Filter heart disease patients only (assuming 1 = heart disease)
df_hd = df[df['Heartdisease'] == 1].copy()

# Standardize Gender labels
df_hd['Gender'] = df_hd['Gender'].str.strip().str.capitalize()
df_hd['Gender'] = df_hd['Gender'].replace({'M': 'Male', 'F': 'Female'})

# Create color map for genders
color_map = {'Male': 'purple', 'Female': 'red'}  # blue & orange

plt.figure(figsize=(8, 6))

# Scatter plot: RBC (x-axis) with jittered y-axis
for gender, color in color_map.items():
    subset = df_hd[df_hd['Gender'] == gender]
    y_jitter = np.random.normal(loc=0, scale=0.05, size=len(subset))  # smaller jitter for neatness
    plt.scatter(subset['RBC'], y_jitter, label=gender,
                color=color, alpha=0.7, edgecolor='k', s=60)

# Formatting
plt.yticks([])  # Hide y-axis (purely for jitter separation)
plt.xlabel('Red Blood Cell Count (RBC) ($10^{12}$/L)', fontsize=14)
plt.title('RBC Distribution in Heart Disease Patients by Gender', fontsize=16, fontweight='bold')
plt.legend(title='Gender')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xlim(df_hd['RBC'].min() - 0.5, df_hd['RBC'].max() + 0.5)  # buffer for edges

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('actualdata.csv')

# Filter non-heart disease patients only
df_no_hd = df[df['Heartdisease'] == 0].copy()

# Standardize Gender labels
df_no_hd['Gender'] = df_no_hd['Gender'].str.strip().str.capitalize()
df_no_hd['Gender'] = df_no_hd['Gender'].replace({'M': 'Male', 'F': 'Female'})

# Count gender occurrences
gender_counts = df_no_hd['Gender'].value_counts()

# Plot bar chart
plt.figure(figsize=(6, 5))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette=['purple', 'red'],width=0.2)
plt.title('Gender Distribution of Non-Heart Disease Patients', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
