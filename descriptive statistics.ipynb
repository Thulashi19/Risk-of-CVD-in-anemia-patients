```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("actualdata.csv")  # Replace with your actual file path

# Calculate summary stats: mean, std dev, count (N)
numeric_cols = df.select_dtypes(include='number').columns

summary_df = pd.DataFrame({
    'Variable': numeric_cols,
    'Mean': df[numeric_cols].mean().round(2).values,
    'Std. Dev.': df[numeric_cols].std().round(2).values,
    'N': df[numeric_cols].count().values
})

# Print summary table
print("Descriptive Statistics of Key Variables:")
print(summary_df.to_string(index=False))

# Show basic summary statistics (optional)
summary = df.describe()
print("\nSummary Statistics:\n", summary)

# Calculate and print medians separately (optional)
medians = df.median(numeric_only=True)
print("\nMedians:\n", medians)

# Plot histograms for all numerical columns
df.hist(bins=20, figsize=(15, 10), edgecolor='black')
plt.suptitle('Histograms of All Numerical Features')
plt.tight_layout()
plt.show()

# Plot box plots for all numerical columns
plt.figure(figsize=(15, 8))
sns.boxplot(data=df)
plt.title('Box Plots of All Numerical Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
